```python
import numpy as np
from PIL import Image
import urllib.request
import matplotlib.pyplot as plt

# URLから画像を読み込む
url = 'https://github.com/ipython-books/cookbook-2nd-data/blob/master/beach.png?raw=true'
with urllib.request.urlopen(url) as response:
    img = Image.open(response)
    img = np.array(img)[::4, ::4, :]  # 間引き処理

# 表示
plt.imshow(img)
plt.axis('off')
plt.show()
```
